```python
import pandas as pd
X = pd.DataFrame([2 , 6 , 5, 12]).values
y = pd.DataFrame([10 , 13 , 15, 16]).values
```

```python
n = len(X)

m = (n * sum(X * y) - sum(X) * sum(y)) / (n * sum(X**2) - (sum(X))**2)
b = (sum(y) - m * sum(X)) / n

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
import matplotlib.pyplot as plt
pred_y = m*X + b
plt.scatter(X,y)
plt.plot(X, pred_y, color = 'red')
```

```python
# Let's apply G.D assuming slope is constant = 0.60
# and let's assume the starting value b = 9
m = 0.60
y_pred_asu = ((m * X) + 9).reshape(4)
y_pred_asu
```

```python
plt.scatter(X,y)
plt.plot(X, pred_y, color = 'red', label='OLS')
plt.plot(X, y_pred_asu, color = 'green' , label ="b=0")
plt.legend()
plt.show()
```

```python
# loss function
m = 0.60
b = 0

loss_scope = -2 * sum(y-m*X - b)
loss_scope
```

```python
lr = 0.1
step_size = loss_scope * lr
step_size
```

```python
# calculating step_size
b = b - step_size
b
```

```python
y_pred_1 =  ((m * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X, pred_y, color = 'red', label='OLS')
plt.plot(X, y_pred_asu, color = 'green' , label ="b=0")
plt.plot(X, y_pred_1, color = 'yellow' , label ="b=new step size")
plt.legend()
plt.show()
```

```python
lr = 0.0001
epochs = 100

for i in range(epochs):
    
    #
    loss_slope = -2 * sum(y - m * X - b)
    b = b - lr * loss_slope
    
    y_pred = m * X + b

    # draw every line
    plt.plot(X, y_pred, color=plt.cm.Blues(i/epochs), alpha=0.5)

# actual points
plt.scatter(X, y, color='red', s=50)
```
